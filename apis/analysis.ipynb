# API Analysis

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

pd.options.display.float_format = '{:.4f}'.format
```

## Load Data

```python
observations = pd.read_csv(r'observations.csv', header=None)
observations = observations.T
observations.columns = observations.iloc[0]
observations = observations[1:]
observations.head()
```

## Box Plots

```python
fig, ax = plt.subplots(figsize=(12, 7))
# Remove top and right border
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
# Remove y-axis tick marks
ax.yaxis.set_ticks_position('none')
# Add major gridlines in the y-axis
ax.grid(color='grey', axis='y', linestyle='-', linewidth=0.25, alpha=0.5)
# Set plot title
ax.set_title('Response times between different APIs in nano seconds')
# Set species names as labels for the boxplot
dataset = [observations['RPC'].values, observations['REST'].values, observations['GraphQL'].values]
labels = ['RPC', 'REST', 'GraphQL']
ax.boxplot(dataset, labels=labels, showfliers=False)
plt.show()
fig.savefig(f'figures/box-plot.png')
```

```python
figure = plt.figure()
for (columnName, columnData) in observations.iteritems():
    plt.boxplot(columnData.values)

plt.show()
figure.savefig(f'figures/{columnName}.png')
```
